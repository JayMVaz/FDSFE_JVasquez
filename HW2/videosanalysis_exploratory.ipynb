```python
for c in ["views", "likes", "dislikes", "comment_count"]:
  ax = videos.hist(column="dislikes")
  thisaxis = ax[0][0]
  thisaxis.set_yscale('log')
  thisaxis.set_xlabel("dislikes")
  thisaxis.set_ylabel("count")
  thisaxis.ticklabel_format(axis='x', scilimits=(0,9))

```
